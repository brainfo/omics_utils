```python
def outliers(boundaries):
    nmax = np.quantile(boundaries.tadsize,.75)+3*(np.quantile(boundaries.tadsize,.5)-np.quantile(boundaries.tadsize,.25))
    nmin = np.quantile(boundaries.tadsize,.25)-3*(np.quantile(boundaries.tadsize,.5)-np.quantile(boundaries.tadsize,.25))
    out = boundaries[(boundaries['tadsize']>nmax) | (boundaries['tadsize']<nmin)]
    return out
def norm(boundaries):
    nmax = np.quantile(boundaries.tadsize,.75)+3*(np.quantile(boundaries.tadsize,.5)-np.quantile(boundaries.tadsize,.25))
    nmin = np.quantile(boundaries.tadsize,.25)-3*(np.quantile(boundaries.tadsize,.5)-np.quantile(boundaries.tadsize,.25))
    normal = boundaries[(boundaries['tadsize']<=nmax) & (boundaries['tadsize']>=nmin)]
    return normal
```
