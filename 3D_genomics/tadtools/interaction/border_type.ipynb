```python
import pandas as pd
import plotly.graph_objects as go
import plotly.io as pio
pio.kaleido.scope.default_format = "pdf"
import plotly.express as px
from matplotlib.colors import ListedColormap
import matplotlib.colors as mc
import matplotlib.pyplot as plt
from plotly.offline import plot
import numpy as np
## using seaborn
import seaborn as sns
from collections import Counter
```

```python
ab_annotate = pd.read_csv('./data/AB_borders_annotate.bed',sep='\t',header=None,usecols=[3,4,5,6],names=['CTCF','EnhD','EnhP','Prom'])
mb_annotate = pd.read_csv('./data/MB_borders_annotate.bed',sep='\t',header=None,usecols=[3,4,5,6],names=['CTCF','EnhD','EnhP','Prom'])
sb_annotate = pd.read_csv('./data/SB_borders_annotate.bed',sep='\t',header=None,usecols=[3,4,5,6],names=['CTCF','EnhD','EnhP','Prom'])
```

```python
data = pd.concat([ab_annotate,mb_annotate,sb_annotate],ignore_index=True)
data['type'] = ['Type AB']*len(ab_annotate.index)+['Type MB']*len(mb_annotate.index)+['Type SB']*len(sb_annotate.index)
```

```python
hue_order = ["Type AB", "Type MB", "Type SB"]
```

```python
sns.histplot(
    data=data, x="CTCF", hue="type",
    hue_order=hue_order,
    log_scale=False, element="step", fill=False,
    cumulative=True, stat="density", common_norm=False,
)
plt.savefig('./results/CTCF_sns.pdf',bbox_inches='tight',family='sans')
```

```python
SB_repeat = pd.read_csv('./data/SB_borders_repeat.bed',sep='\t',header=None)
MB_repeat = pd.read_csv('./data/MB_borders_repeat.bed',sep='\t',header=None)
AB_repeat = pd.read_csv('./data/AB_borders_repeat.bed',sep='\t',header=None)
```

```python
def to_countdf(repeat):
    repeat_count = Counter(repeat.iloc[:,3])
    df = pd.DataFrame.from_dict(repeat_count, orient='index').reset_index()
    df = df.rename(columns={'index':'Repeat', 0:'Count'})
    return df
```

```python
count_list = list(map(to_countdf,[AB_repeat,MB_repeat,SB_repeat]))
count_df = pd.concat(count_list,axis=1,keys='Repeat',join='outer')
```

```python
count_df = count_df.iloc[:,[0,1,3,5]]
count_df.columns = count_df.columns.droplevel()
count_df.columns = ['Repeat','AB','MB','SB']
count_df.fillna(0,inplace=True)
count_df
```

```python
all_count = {'AB':1124,'MB':238,'SB':1380}
ratio = list(map(lambda x: count_df[x].div(all_count[x]),['AB','MB','SB']))
ratio_df = pd.DataFrame()
ratio_df['Repeat'] = count_df['Repeat']
ratio_df['AB'] = ratio[0]
ratio_df['MB'] = ratio[1]
ratio_df['SB'] = ratio[2]
```
