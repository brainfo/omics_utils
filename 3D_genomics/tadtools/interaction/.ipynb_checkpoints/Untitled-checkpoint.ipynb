```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

```python
stats = pd.read_csv('../10Model/emissions_10.txt',sep='\t',index_col=0)
```

```python
stats =stats*100
```

```python
st = sns.heatmap(stats,annot=True,cmap='YlGnBu')
stfig = st.get_figure()
```

```python
stfig.savefig('10States.pdf',bbox_inches='tight')
```
