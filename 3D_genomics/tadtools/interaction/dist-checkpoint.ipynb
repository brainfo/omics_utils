```python
from os import listdir
import os
import pandas as pd
import numpy as np
cwd = os.getcwd()
dir_rep1 = cwd + '/rep1'
dir_wt2 = cwd +'/wt2'
dir_rep2 = cwd + '/rep2'
dir_wt1 = cwd + '/wt1'
dir_ren = cwd + '/ren'
ch = list(range(1,7,1))
ch += list(range(8,20,1))
ch += ['X']
```

```python
brep1 = pd.read_csv('./rep1/rep1-bound.txt',sep='\t')
brep2 = pd.read_csv('./rep2/rep2-bound.txt',sep='\t')
bwt1 = pd.read_csv('./wt1/wt1-bound.txt',sep='\t')
bren = pd.read_csv('./ren/ren-bound.txt',sep='\t')
```

```python
def distance(my,other):
    other = np.asarray(other)
    dist = abs(my-other)
    min_dist = min(dist)
    return min_dist
```

```python
def perchr(data,chrom):
    data_chrom = data.loc[data.chrom == chrom]
    return data_chrom
```

```python
dist = []
for c in ch:
    chrom = 'chr%s' %(c)
    wt1data = np.asarray(perchr(bwt1,chrom)['start'])
    rep1data = perchr(brep1,chrom)['start']
    rep2data = perchr(brep2,chrom)['start']
    rendata = perchr(bren,chrom)['start']
    fun = lambda t: distance(t,rep1data)
    dist += list(np.array([fun(xi) for xi in wt1data]))
    fun = lambda t: distance(t,rep2data)
    dist += list(np.array([fun(xi) for xi in wt1data]))
    fun = lambda t: distance(t,rendata)
    dist += list(np.array([fun(xi) for xi in wt1data]))
```
